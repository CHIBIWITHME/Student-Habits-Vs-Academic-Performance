CORRELATION HEATMAP (NUMERICAL FEATURES)

```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include='number')

# Correlation matrix
corr_matrix = numeric_cols.corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix')
plt.show()
```

Study hours per day strongly correlates with exam score. Others are mental health rating, exercise frequency, sleep hours, and attendance percentage which are all positive correlations to exam score.

## Insights Summary from EDA & Bivariate Analysis

---

### Performance Overview
- Most students scored between **60â€“80**.
- A few students scored above 90 (high performers); some scored below 50 (struggling).

- **Study Hours per Day**: A clear **positive correlation** with exam scores. Students studying **3â€“5 hours/day** tend to perform better.

- **Netflix Hours**: Similar to social media â€” excessive time correlates with slightly lower scores.
- **Attendance Percentage**: Strong positive correlation â€” higher attendance typically results in higher exam scores.

## Data-Driven Recommendations to Improve Student Performance

Based on the data analysis and observed trends, here are actionable recommendations to help students improve their academic performance:

---

### 1. Optimize Study Habits
- Encourage students to **study at least 3â€“5 hours/day** for optimal results.
- Emphasize **consistency over cramming** â€” regular study hours correlate with higher exam scores.

---

### 2. Prioritize Healthy Sleep
- Promote **6.5 to 8 hours of sleep** per night.
- Highlight that inadequate sleep (<5 hours) is associated with **significantly lower performance**.

---

### 3. Manage Screen Time Wisely
- Students spending **more than 3 hours/day** on social media or streaming platforms (e.g., Netflix) tend to underperform.
- Schools can offer **digital well-being workshops** or introduce app usage tracking tools.

---

### 4. Improve Class Attendance
- Reinforce policies that encourage students to maintain **80%+ attendance**.
- Consider implementing **attendance incentives** or feedback systems.

---

### 5. Promote Better Diet and Lifestyle
- Students reporting **Good/Excellent diet quality** showed higher scores.
- Schools can run **nutrition awareness programs** and promote **on-campus healthy food options**.

---

### 6. Support Mental Health
- A higher mental health rating correlates with better scores.
- Recommendations:
  - Increase access to **counselors and mental health resources**.
  - Encourage **moderate exercise** (2â€“4 times/week).
  - Host **stress-relief workshops** during exam season.

---

### 7. Improve Access to Internet & Learning Tools
- Students with **good internet quality** perform better â€” likely due to improved access to online materials.
- Suggest partnerships with ISPs or tech organizations to provide **low-cost access and devices**.

---

### 8. Parental Engagement
- Students with more educated parents tend to score better.
- Schools could:
  - Organize **parental guidance sessions** on supporting children's academic habits.
  - Share **monthly academic summaries** with guardians.

---

> These targeted interventions can help schools and students create a healthier, more effective learning environment â€” improving both academic outcomes and overall well-being.
